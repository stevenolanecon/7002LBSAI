```python
import statsmodels.formula.api as smf
```

```python
import pandas as pd
```

```python
file_path = 'synthetic_reaction_times.csv'
df = pd.read_csv(file_path)
display(df.head())
```

```python
model = smf.ols('reaction_time_ms ~ finger_mm + caffeine + device', data=df).fit(cov_type='HC1')
print(model.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Get the exogenous variables (independent variables) from the fitted model
exog = model.model.exog
exog_df = pd.DataFrame(exog, columns=model.model.exog_names)

# Calculate VIF for each independent variable
vif_data = pd.DataFrame()
vif_data['feature'] = exog_df.columns
vif_data['VIF'] = [variance_inflation_factor(exog_df.values, i) for i in range(exog_df.shape[1])]

print(vif_data)
```
